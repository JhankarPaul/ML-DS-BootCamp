```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df=sns.load_dataset("tips")
df.head()
```

```python
df.shape
```

```python
df.corr()
```

```python
sns.heatmap(df.corr())
```

# Pair Plot

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='time')
```

# Distribution Plot (checking distribution of the features)

```python
sns.distplot(df["tip"])
```

```python
sns.displot(df["tip"])
```

```python
sns.histplot(df["tip"])
```

# Jointplot

```python
sns.jointplot(x='size',y='total_bill',data=df)
```

```python
sns.jointplot(x='tip',y='total_bill',data=df)
```

```python
sns.jointplot(x='size',y='tip',data=df,kind='red')
```

```python
sns.jointplot(x='size',y='tip',data=df,kind='kde')
```

```python
sns.jointplot(x='size',y='tip',data=df,kind='hist')
```

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='hex')
```

# CountPlot

```python
sns.countplot(data=df,x='sex')
```

```python
sns.countplot(data=df,y='sex')
```

# Bar Plot

```python
sns.barplot(x='sex',y='total_bill',data=df)
```

```python
sns.barplot(x='sex',y='tip',data=df)
```

# Box Plot

```python
sns.boxplot(data=df,x='day',y='total_bill')
```

```python
sns.boxplot(data=df,x='day',y='total_bill',hue='smoker')
```

# Violin Plot (based on density distribution)

```python
sns.violinplot(x='day',y='total_bill',data=df)
```
