# Subplot     (2,2,1 means nrows,ncols,index)

```python
p=np.arange(1,6)
q=p
r=p**2
s=p**3
t=p**4
print(p,q,r,s,t)
```

```python
plt.subplot(2,2,1)
plt.plot(p,q,'r.--')
plt.title('fig.1')
plt.subplot(2,2,2)
plt.plot(p,r,'g--')
plt.title('fig.2')
plt.subplot(2,2,3)
plt.plot(p,s,'b--')
plt.title('fig.3')
plt.subplot(2,2,4)
plt.plot(p,t,'p--')
plt.title('fig.4')
```

```python
x=np.arange(0,6*np.pi,0.1)
y=np.sin(x)
plt.plot(x,y)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Sine Curve')
```

# BAR PLOT

```python
x=[1,2,4,9]
y=[25,4,16,81]
plt.bar(x,y)
plt.show()
```

```python
x=[1,2,4,9]
y=[25,7,16,8]

y1=[1,2,4,9]
x1=[25,7,16,8]

plt.bar(x,y,color='r')
plt.bar(x1,y1,color='g')

plt.title('Bar Graph')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.show()
```

# PIE Chart

```python
labels=['python','java','c++','c']
sizes=[215,313,145,200]
colors=['gold','yellowgreen','lightcoral','lightskyblue']
explode=(0.4,0,0,0)

plt.pie(sizes,explode=explode,labels=labels,colors=colors,autopct='%1.1f%%',shadow=False)

plt.axis('equal')
plt.show()
```

# BOXPLOT

```python
data=[np.random.normal(0,std,200) for std in range(1,4)]
#rectangular box plot
plt.boxplot(data,vert=True,patch_artist=False)
plt.show()
```
